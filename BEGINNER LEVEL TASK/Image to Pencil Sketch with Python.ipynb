# Image to Pencil Sketch with Python

```python
# import a cv2 library
import cv2
```

```python
# import matplotlib 
import matplotlib.pyplot as plt
# pyplot is aliased as plt
```

### Upload an image

```python
# uploding an image
img = cv2.imread('C:/Users/Admin/Desktop/Dhoni.jpg')
```

```python
# displaying a original image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()
```

### convert into grayscale image

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_image)
plt.title('Gray image')
plt.show()
```

### changing in inverted gray image

```python
inverted_gray_image = 255 - gray_image
```

```python
# displaying a inverted image
plt.imshow(inverted_gray_image)
plt.title('Inverted Image')
plt.show()
```

### Blurring an image

```python
#blurring
blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)
```

```python
# displaying a blurred image
plt.imshow(blurred_image)
plt.title('Blurred Image')
plt.show()
```

### convert into pencil sketch image

```python
pencil_sketch_image = cv2.divide(gray_image, 255 - blurred_image, scale=256.0)
```

```python
#pencil_sketch_image = cv2.multiply(pencil_sketch_image, 0.8)  # multiply with 0.8 to make it a little more gray if u want
```

```python
# displaying a pencil sketch image
plt.imshow(pencil_sketch_image,cmap="gray")
plt.title('Pencil Sketch Image')
plt.show()
```

### Showing a original image and pencil sketch image

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[1].imshow(pencil_sketch_image, cmap='gray')
axs[1].set_title('Pencil Sketch Image')
plt.show()
```
