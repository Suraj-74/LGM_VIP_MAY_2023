Handling null values

```python
# filll the null vaslues of numerical column by taking median
cols=["kill","wound"]
median_val=df1[cols].median()
df1[cols]=df1[cols].fillna(median_val)
```

```python
df1.isnull().sum()  # checking the null values 
```

```python
# now, we are going to handle the null value of cartegorical column by using mode
cols=["state","city","summary","motive"]
mode=df1[cols].mode().iloc[0]
df1[cols]=df1[cols].fillna(mode)
```

```python
# Again check the null values in the dataset
df1.isnull().sum()
```

Now, there are no null values in the dataset

```python
# Treating Null Values in State Column
df1['state'] = df1['state'].replace({'Unknown': np.nan})
df1['state'] = df1['state'].fillna("Others")
df1['state'].value_counts()
```

```python
# Checking for null values
df1['state'].isna().any()
```

```python
# Treating Null Values in Target Column
df1['Targettype'] = df1['Targettype'].replace({'Unknown': np.nan})
df1['Targettype'] = df1['Targettype'].fillna("Others")
df1['Targettype'].value_counts()
```

```python
# Checking for null values
df1['Targettype'].isnull().any()
```
